# Regresión

- También se puede realizar regresión con árboles de decisión.
- La salida del árbol será el promedio de los valores de las observaciones que caigan en la hoja final.
- La función de costo en este caso puede ser el MSE o el MAE.
- El resto de los parámetros analizados para clasificación siguen siendo válidos.


```python
print(__doc__)

# Import the necessary modules and libraries
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
import matplotlib.pyplot as plt

# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=15)
regr_3 = RandomForestRegressor(max_depth=5)
regr_4 = ExtraTreesRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
regr_3.fit(X, y)
regr_4.fit(X, y)

# Predict
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
y_3 = regr_3.predict(X_test)
y_4 = regr_4.predict(X_test)

# Plot the results
plt.figure(figsize=[12,12])
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",
         label="DT max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="DT max_depth=15", linewidth=2)
plt.plot(X_test, y_3, color="blue", label="RandomForest", linewidth=2)
plt.plot(X_test, y_4, color="green", label="Extra Trees", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
